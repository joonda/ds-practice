## data preprocessing

### crime

```python
crime.head()
```

```python
crime['연령'].value_counts()
```

```python
drop_idx = crime[(crime['연령'].str.contains('이하')) | (crime['연령'].str.contains('미상'))].index
crime.drop(index=drop_idx, inplace=True)
crime
```

```python
station_list = []
crime_sort_list = []
for idx, row in crime.iterrows():
    station = row['구분'].split()[0]
    station_list.append(station)
    crime_sort = row['구분'].split()[1]
    crime_sort_list.append(crime_sort)

crime_sort_list[:5], station_list[:5]
```

```python
crime['구분'] = station_list
crime['죄종'] = crime_sort_list

crime.head()
```

```python
crime_station = crime.groupby(['구분', '죄종'])["피해자 수"].sum().to_frame().reset_index()
crime_station.head()
```

```python
crime_station = crime_station.pivot(index=['구분'], columns=['죄종'], values='피해자 수').reset_index()
crime_station.head()
```

### population

```python
pop = pd.read_excel('./data/자치구별+고령인구(추계인구)_20241019165325.xlsx')
pop.head()
```

```python
pop.drop(index=[0, 1, 2], columns=['자치구별(1)'], inplace=True)
pop.columns = ["구별", "노령인구_전체", "노령인구_남자", "노령인구_여자"]
pop.reset_index(drop=True, inplace=True)
```

### old-point

```python
old_point = pd.read_excel('./data/자치구별+노령화지수(추계인구)_20240929181758.xlsx')
old_point.head()
```

### concat

```python
cctv_sum.head()
```

```python
crime_station.head()
```

```python
area.head()
```

```python
pop.head()
```

```python
old_point.head()
```

```python
merge_crime = pd.merge(crime_station, cctv_sum, left_on='구별', right_on='자치구', how='inner')
merge_crime.drop(columns='자치구', inplace=True)
merge_crime.head()
```

```python
merge_crime = pd.merge(merge_crime, area, on='구별')
merge_crime.head()
```

```python
merge_crime = pd.merge(merge_crime, pop, on='구별')
merge_crime.head()
```

```python
merge_crime = pd.merge(merge_crime, old_point, on='구별')
merge_crime.head()
```

```python
merge_crime.info()
```

```python
merge_crime['면적 (km²)'] = merge_crime['면적 (km²)'].astype(float) 
merge_crime['구성비 (%)'] = merge_crime['구성비 (%)'].astype(float)
merge_crime['전체'] = merge_crime['전체'].astype(int)
merge_crime['남자'] = merge_crime['남자'].astype(int)
merge_crime['여자'] = merge_crime['여자'].astype(int)
```

```python
merge_crime['면적 당 CCTV 대수(km² 당)'] = round((merge_crime['CCTV 수량'] / merge_crime['면적 (km²)']), 2)
merge_crime.head()
```

```python
merge_crime['노령인구 당 CCTV 비율(인당) (%)'] = round((merge_crime['CCTV 수량'] / merge_crime['전체']) * 100, 2)
merge_crime.head()
```

## data visualization

### 구 면적 당 CCTV 대수

```python
merge_crime_sorted = merge_crime.sort_values(by='면적 당 CCTV 대수(km² 당)', ascending=False)

plt.figure(figsize=(18, 10))
ax = sns.barplot(data=merge_crime_sorted, x='구별', y='면적 당 CCTV 대수(km² 당)', palette='viridis', hue='구별')
plt.title("구 면적 당 CCTV 대수")
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', 
                fontsize=10, color='black', 
                xytext=(0, 5), 
                textcoords='offset points')
plt.xticks(rotation=45)
plt.show()
```

* 성동구, 광진구, 중구, 양천구, 금천구 순으로 구 면적당 CCTV 대수가 높다

### 노령인구 당 CCTV 비율

```python
merge_crime_sorted = merge_crime.sort_values(by='노령인구 당 CCTV 비율(인당) (%)', ascending=False)

plt.figure(figsize=(18, 10))
ax = sns.barplot(data=merge_crime_sorted, x='구별', y='노령인구 당 CCTV 비율(인당) (%)', palette='viridis', hue='구별')
plt.title("노령인구 당 CCTV 비율(인당) (%)")
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', 
                fontsize=10, color='black', 
                xytext=(0, 5), 
                textcoords='offset points')
plt.xticks(rotation=45)
plt.show()
```

* 하지만, 노령인구 당 CCTV 비율은 중구, 강남구, 성동구, 서초구, 광진구 순

```python
merge_crime_sorted = merge_crime.sort_values(by='전체', ascending=False)

plt.figure(figsize=(18, 10))
ax = sns.barplot(data=merge_crime_sorted, x='구별', y='전체', palette='viridis', hue='구별')
plt.title("구별 노령인구")
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', 
                fontsize=10, color='black', 
                xytext=(0, 5), 
                textcoords='offset points')
plt.xticks(rotation=45)
plt.show()
```
